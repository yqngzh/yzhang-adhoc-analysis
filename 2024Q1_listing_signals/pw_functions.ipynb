```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0.0, 30.0, 0.5)
y_log1p = np.log(1 + x) + 1
y_sqrt = np.sqrt(1 + x)
y_pow_07 = (1 + x) ** 0.7
y_pow_04 = (1 + x) ** 0.4
y_pow_01 = (1 + x) ** 0.1
```

```python
plt.plot(x, y_pow_07, label="pow_07", color="red")
plt.plot(x, y_pow_04, label="pow_04", color="orange")
plt.plot(x, y_pow_01, label="pow_01", color="pink")
plt.plot(x, y_log1p, label="log1p", color="green")
plt.plot(x, y_sqrt, label="sqrt", color="blue")
plt.legend()
plt.show()
```
