```python
si_fi_list = []

with open("./si_raw_importance.txt", "r") as file:
    for line in file:
        if "name" in line:
            line_split = line.split(": ")
            feature_name = line_split[1].replace('"', '').replace(',', '').strip()
            si_fi_list.append(feature_name)
```

```python
len(si_fi_list)
```

```python
so_fi_list = []

with open("./so_raw_importance.txt", "r") as file:
    for line in file:
        if "name" in line:
            line_split = line.split(": ")
            feature_name = line_split[1].replace('"', '').replace(',', '').strip()
            so_fi_list.append(feature_name)
```

```python
len(so_fi_list)
```

```python
si_config_list = []

with open("./si_config.txt", "r") as file:
    for line in file:
        if "name" in line:
            line_split = line.split(": ")
            feature_name = line_split[1].replace("''", '').replace('\n', '').strip()
            si_config_list.append(feature_name)
```

```python
len(si_config_list)
```

```python
so_config_list = []

with open("./so_config.txt", "r") as file:
    for line in file:
        if "name" in line:
            line_split = line.split(": ")
            feature_name = line_split[1].replace("''", '').replace('\n', '').strip()
            so_config_list.append(feature_name)
```

```python
len(so_config_list)
```

```python
full_feature_list = si_config_list + so_config_list
full_feature_list = sorted(list(set(full_feature_list)))
full_feature_list
```
